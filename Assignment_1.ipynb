# 5. Perform meaningful visualizations using the dataset.

```python
#1.importing visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

# Distplot

```python
#distplot-- distribution of sepallength
sns.distplot(df.SepalLengthCm,bins=20,color='r')
```

# Boxplot

```python
sns.boxplot(x='SepalWidthCm',data=df)
```

There are outliers in sepalwidthcm

# Lineplot

```python
sns.lineplot(x='PetalLengthCm',y='PetalWidthCm',data=df)
```

petallength increses with petalwidth

# Scatterplot

```python
sns.scatterplot(x='SepalLengthCm',y='PetalLengthCm',hue='Species',data=df)
```

iris verginica has large petallength and sepallength

# Barplot

```python
sns.barplot(x='Species',y='SepalWidthCm',data=df)
```

iris setosa has largest sepalwidth compared to versicolor and virginica

# Heatmap

```python
sns.heatmap(data=df.corr(),annot=True)
```
